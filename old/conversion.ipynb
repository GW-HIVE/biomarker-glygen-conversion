Split up literature evidence.

```python
def split_pmid(text):
    results = []
    if "[PMID:" in text:
        parts = re.split(r"(\[PMID:\s*\d+\])", text)
        for i in range(0, len(parts) - 1, 2):
            if i + 1 < len(parts):
                results.append(parts[i].strip() + " " + parts[i + 1])
    else:
        results.append(text)
    return results
glygen_df["split_evidence"] = glygen_df["Literature evidence"].apply(split_pmid)
glygen_df["split_evidence"].iloc[1]
```

```python
def split_evidence(text):
    if "PMID:" in text:
        parts = text.split("[PMID:", 1)
        if len(parts) == 2:
            return [parts[0].strip(), "Pubmed:" + parts[1].strip()]
    return [text, "None"]
        
glygen_df[["evidence", "evidence_source"]] = glygen_df["split_evidence"].apply(split_evidence).tolist()
glygen_df["evidence_source"] = glygen_df["evidence_source"].str.replace("]", "")
glygen_df.head()
```

```python
def clean_parantheticals(source_column: pd.Series) -> pd.Series:
    """Cleans the parantheticals in the given dataframe column if the
    value is a string.

    Parameters
    ----------
    source_column: pd.Series
        The dataframe column to clean.

    Returns
    -------
    pd.Series
        The cleaned dataframe column.
    """
    return source_column.apply(
        lambda x: re.sub(r"\([^)]*\)", "", x).strip() if isinstance(x, str) else x
    )

updated_df["assessed_biomarker_entity"] = clean_parantheticals(updated_df["assessed_biomarker_entity"])
updated_df.head()
```
